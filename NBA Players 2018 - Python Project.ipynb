```python
import pandas as pd
df = pd.read_csv("nba2018.csv")
print(df)
```

```python
                                    ######## CLEANING UP THE DATA ########
# shows first 5 rows as default

# displays all columns in the data
pd.set_option('display.max_columns', None)

# displays data in a nice table format 
display(df.head())

# removes all null 'NaN' (not a number) values
df.dropna(inplace=True)

# changes column name from "number" to "jersey number"
df.rename(columns = {'Number':'Jersey Number'}, inplace = True)

# changes float numbers into integers for age, number, and salary columns
df['Jersey Number'] = df['Jersey Number'].astype(int)

df['Age'] = df['Age'].astype(int)

df['Salary'] = df['Salary'].astype(int)

# sorts values in salary column in ascending order
df.sort_values(by=['Salary'], inplace=True, ascending=False)
```

```python
# basic descriptive statistics on a dataframe
df['Salary'].describe()
```

```python
                ######## Question 1: Which players make over 15 million in salary? ########

# locates the 'salary' column and lists players who salaries are greater than or equal to 15 million
df.loc[df['Salary'] >= 15000000]
```

```python
            ######## Question 2: Highest paid player on the Dallas Mavericks roster? ########

# groups by team column
df = df.groupby('Team')

# specifies to only show players on the dallas mavericks team
df.get_group('Dallas Mavericks')
```
